## Dummy Variables
##### var_list=['season', 'yr', 'mnth', 'holiday', 'weekday','workingday','weathersit']
##### We will assign dummy variables to season,mnth,weekday,weathersit
###### yr,holiday,workingday already has encoding 0 and 1

```python
var_list_dummy=['season','mnth','weekday','weathersit']
```

```python
season=pd.get_dummies(bike_df['season'],drop_first='True')
season.head()
```

```python
season.set_axis(['summer', 'fall', 'winter'], axis='columns', inplace=True)
```

```python
season.info()
season.head(100)
```

```python
bike_df=pd.concat([bike_df,season],axis=1)
bike_df.head()
```

```python
month=pd.get_dummies(bike_df['mnth'],drop_first='True')
month.head()
```

```python
month.set_axis(['february', 'march', 'april','may','june','july','august','september','october','november','december'], axis='columns', inplace=True)
month.head()
```

```python
bike_df=pd.concat([bike_df,month],axis=1)
bike_df.head()
```

```python
weekday=pd.get_dummies(bike_df['weekday'],drop_first='True')
weekday.set_axis(['tuesday', 'wednesday', 'thursday','friday','saturday','sunday'], axis='columns', inplace=True)
bike_df=pd.concat([bike_df,weekday],axis=1)
```

```python
bike_df.head()
```

```python
weather=pd.get_dummies(bike_df['weathersit'],drop_first='True')
weather.set_axis(['misty', 'light_snow'], axis='columns', inplace=True)
bike_df=pd.concat([bike_df,weather],axis=1)
```

```python
bike_df.head()
```

```python
# Now deleting all the  columns for which encoding has been generated
bike_df=bike_df.drop(columns=['season','mnth','weekday','weathersit'],axis=1)
bike_df.head()
```

```python
# Now deleting the unnecessary columns
bike_df=bike_df.drop(columns=['atemp','casual','registered'],axis=1)
bike_df.head()
```

Finally the data has been prepared. 

# Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively

bike_train, bike_test = train_test_split(bike_df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

# Rescaling the Features 

We will use MinMax scaling.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['temp', 'hum', 'windspeed', 'cnt']

bike_train[num_vars] = scaler.fit_transform(bike_train[num_vars])

bike_train.head()
```

### Dividing into X and Y sets for the model building

```python
y_train = bike_train.pop('cnt')
X_train = bike_train
```

## Building the model using RFE

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 12
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, step=12)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

### Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

We can see that the adjusted R squared is 77% which is not bad . 
We will see if we can improve this value by removing any other variable.

```python
# Removing thursday as it has a very high p value
X_train_new = X_train_rfe.drop(["thursday"], axis = 1)
```

```python
import statsmodels.api as sm 
X_train_lm = sm.add_constant(X_train_new)
lm = sm.OLS(y_train,X_train_lm).fit() 
```

```python
print(lm.summary())
```

We can see the r-squared has significantly no difference so let us look at the variance too.

```python
X_train_new.columns
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

Only const has a high VIF rest for all it is less than 5 , so we don't need to remove these variables now and let's keep this model as it is.

But looking at the p values few more working days can be deleted but we are keeping this model like this 

## Residual Analysis of the train data
So, now to check if the error terms are also normally distributed (which is infact, one of the major assumptions of linear regression), let us plot the histogram of the error terms and see what it looks like.

```python
y_train_cnt = lm.predict(X_train_lm)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

This looks more or less like a normal distribution with the histogram skewed a little bit more to the left side but that is fine.
The error terms seem to be centred around 0 which is the basic assumption of Linear regression.

## Making Predictions

```python
num_vars = ['temp', 'hum', 'windspeed', 'cnt']

bike_test[num_vars] = scaler.transform(bike_test[num_vars])
```

```python
y_test = bike_test.pop('cnt')
X_test = bike_test
```

```python
# Now let's use our model to make predictions.

# Creating X_test_new dataframe by dropping variables from X_test
X_train_new=X_train_new.drop(columns=['const'],axis=1)
X_test_new = X_test[X_train_new.columns]

# Adding a constant variable 
X_test_new = sm.add_constant(X_test_new)
```

```python
# Making predictions
y_pred = lm.predict(X_test_new)
```

## Model Evaluation

```python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16) 
```

### Calculating the r2 score

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
r2_score(y_true=y_test,y_pred=y_pred)
```

The value is about 0.7623 which is 76% and it is quite a good value almost close to the training one.

```python
mean_squared_error(y_true=y_test,y_pred=y_pred)
```

This value is close to 0 which indicates that this is a good predictor.
